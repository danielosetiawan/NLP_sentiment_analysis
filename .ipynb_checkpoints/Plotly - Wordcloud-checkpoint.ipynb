```python
company='AAPL'
df_comp = cleaned_df[cleaned_df['company'] == company]
# filter by sentiments
df_comp_bull = df_comp[df_comp['sentiment'] == 1]
df_comp_bear = df_comp[df_comp['sentiment'] == 0] 
text_bull = ' '.join(i for i in df_comp_bull['body'])
text_bear = ' '.join(i for i in df_comp_bear['body'])

from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# load two images
image1 = np.array(Image.open("./logos/bull.png"))
image2 = np.array(Image.open("./logos/bear.png"))

# create a WordCloud object with the red image
wordcloud1 = WordCloud(background_color="white", mask=image1, colormap="Reds").generate(text_bull)

# create a WordCloud object with the green image
wordcloud2 = WordCloud(background_color="white", mask=image2, colormap="Greens").generate(text_bear)

# create an ImageColorGenerator for the red image
color_func1 = ImageColorGenerator(image1)

# create an ImageColorGenerator for the green image
color_func2 = ImageColorGenerator(image2)

# apply the color functions to the WordCloud objects
wordcloud1 = wordcloud1.recolor(color_func=color_func1)
wordcloud2 = wordcloud2.recolor(color_func=color_func2)

# display the WordClouds side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(wordcloud1, interpolation='bilinear')
ax1.axis("off")
ax2.imshow(wordcloud2, interpolation='bilinear')
ax2.axis("off")
plt.show()
```
