Use SentimentIntensityAnalyzer() to generate scores/labels as ground truth. Train neural networks to classify Twitter sentiments and compare to generated labels. Stock ticker tweets from 2015-2022

```python
tweets = pd.read_csv('stock_tweets.csv', parse_dates=['Date'])
stocks = pd.read_csv('stock_yfinance_data.csv')
co_values = pd.read_csv('CompanyValues.csv')
co_tweets = pd.read_csv('Company_Tweet.csv')
Tweet = pd.read_csv('Tweet.csv')
scraped = pd.read_csv('scraped_tweets_2014_2019.csv', lineterminator='\n')
scraped2 = pd.read_csv('scraped_tweets_2020_2021.csv', lineterminator='\n')
```

```python
scraped_combined = pd.DataFrame(np.vstack((scraped2.values, scraped.values)))
scraped_combined = scraped_combined[scraped_combined.columns[:-1]]
scraped_combined.columns =['Date', 'Content', 'ticker']
scraped_combined
```

```python
Tweet
```

```python
from datetime import datetime
Tweet['post_date'] = pd.to_datetime(Tweet['post_date'],unit='s')
```

```python
co_tweets
```

```python
co_tweets = co_tweets.replace(to_replace="GOOGL", value="GOOG")
```

```python
combined = pd.merge(Tweet, co_tweets, on=["tweet_id"])
combined
```

```python
df1 = combined[['post_date', 'body', 'ticker_symbol']]
df1
```

```python
df2 = tweets[['Date', 'Tweet', 'Stock Name']]
df2
```

```python
new_frame = pd.DataFrame(np.vstack((df1.values, df2.values)))
new_frame.columns =['Date', 'Content', 'ticker']
new_frame
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

new_frame['score'] = new_frame['Content'].apply(lambda x: sia.polarity_scores(x)['compound'])
# create label
new_frame['label'] = pd.cut(np.array(new_frame['score']),bins=[-1, -0.66, 0.32, 1], right=True,
                          labels=["negative", "neutral", "positive"])
```

```python
scraped_combined['score'] = scraped_combined['Content'].apply(lambda x: sia.polarity_scores(x)['compound'])
scraped_combined['label'] = pd.cut(np.array(scraped_combined['score']),bins=[-1, -0.66, 0.32, 1], right=True,
                          labels=["negative", "neutral", "positive"])
```

```python
all_df = pd.DataFrame(np.vstack((new_frame.values, scraped_combined.values)))
all_df.columns =['Date', 'Content', 'ticker', 'score', 'label']
all_df
```
